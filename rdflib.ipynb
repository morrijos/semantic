```python
# Use simple example from Getting Started with RDFLib—https://rdflib.readthedocs.io/en/stable/gettingstarted/
from rdflib import Graph

# Create a Graph
g = Graph()

# Parse an RDF file
g.parse("files/wine.rdf", format="xml")

# Loop through each triple in graph (subj, pred, obj)
for subj, pred, obj in g:
    # print(subj, pred, obj)
    # Check if there is at least one triple in Graph
    if (subj, pred, obj) not in g:
       raise Exception("It better be!")
```

```python
# Print number of triples in graph
print(f"Graph g has {len(g)} statements.")
# Print: Graph G has 86 statements
```

```python
# Print out entire Graph in RDF Turtle format
print(g.serialize(format="turtle"))
```

```python
q = """
PREFIX wine: <http://www.w3.org/TR/2003/PR-owl-guide-20031209/wine#>

SELECT ?wine ?maker
WHERE {
    ?wine wine:hasMaker ?maker .
}
LIMIT 10
"""

for row in g.query(q):
    print(row.wine, row.maker)
```

```python
# Write SPARQL query
# This query looks for resources that are types of Wine and gets their labels
query = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX wine: <http://w3.org>

SELECT ?wine ?label
WHERE {
    ?wine rdf:type/rdfs:subClassOf* wine:Wine .
    ?wine rdfs:label ?label .
}
LIMIT 10
"""

# Execute the query
results = g.query(query)

# Print results
for row in results:
    print(f"Wine: {row.wine} | Label: {row.label}")
```
